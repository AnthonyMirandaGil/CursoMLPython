# Solucion Leer Datos desde URL


```python
# Cargamos la libreria y hacemos la conexion con los wed de los datos
def downloadFromURL(url,filename,sep = ",",delin = "\n",mainpath = "/home/administrador/Documentos/ML_Python/datasets"):
    import urllib3
    import os
    http = urllib3.PoolManager()
    r = http.request('GET',url)
    r.status
    response = r.data

    # El objeto response contiene un string binario asi que lo convertmos en string decodifacandolo en utf-8
    str_data = response.decode('utf-8')

    # Dividimos el string en un array de filas.
    lines = str_data.split(delin)

    # la primera linea contiene la cabecera asi que la extraemos
    col_names = lines[0].split(sep)
    n_cols = len(col_names)

    # Generamos un diccionario vacio donde ira la informacion procesada desde la URL externa
    main_dict = {}

    # Procesamo fila a fila donde iremos rellenado la informacion en el diccionario
    for col in col_names:
        main_dict[col] = []

    counter = 0

    for line in lines:
        if(counter > 0):
            value = line.split(",")
            for i in range(n_cols):
                main_dict[col_names[i]].append(value[i])
        counter +=1

    print("El data set contiene %d filas y %d columnas "%(counter,n_cols))

    fullpath = os.path.join(mainpath,filename)

    #Convertimos en diccionario en dataframe
    df = pd.DataFrame(main_dict)

    # Guardamos en diferentes formatos
    df.to_csv(fullpath + ".csv")
    df.to_excel(fullpath + ".xls")
    df.to_json(fullpath + ".json")
    return df
```

```python
medals_url = "http://www.winterolympicsmedals.com/medals.csv"
medals_df = downloadFromURL(medals_url,"athletes/downloaded_medals")  
medals_df.head()
```
